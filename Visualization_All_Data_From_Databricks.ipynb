## Efetuando leitura do Parquet e Criando visão para consultas SQL

```python
#Carregando dados do parquet
covid = spark.read.parquet("/FileStore/covid_data_lake/cleaned/year_register=*/")
```

```python
# Criação de TempView para utilização de linguagem sql
covid.createOrReplaceTempView('covid_cases')
```

## Panorama diário de casos confirmados do país com maior número

```python
#Criando uma view para o pais mais casos de mortes para uma futura analise
df_top_confirmed = spark.sql("select country, max(confirmed) as confirmed from covid_cases group by country order by confirmed desc limit 1")
df_top_confirmed.createOrReplaceTempView('top_confirmed')
```

```python
#Criando o dataframe com os dados do pais com o maior numero de casos
df_confirmed = spark.sql("select covid_cases.* from covid_cases inner join top_confirmed on top_confirmed.country = covid_cases.country order by covid_cases.date_register limit 500")
```

```python
display(df_confirmed)
```

## Panorama diário de mortes do país com maior número

```python
#Criando uma view para o pais mais casos de mortes para uma futura analise
df_top_deaths = spark.sql("select country, max(deaths) as deaths from covid_cases group by country order by deaths desc limit 1")
df_top_deaths.createOrReplaceTempView('top_deaths')
```

```python
#Criando o dataframe com os dados do pais com o maior numero de casos
df_deaths = spark.sql("select covid_cases.* from covid_cases inner join top_deaths on top_deaths.country = covid_cases.country order by covid_cases.date_register limit 500")
```

```python
display(df_deaths)
```
